```python
# Exercise 11: Check Primality Functions

def is_prime(num):
    for i in range(2, int(num**0.5+1)):
        if num%i==0:
            return str(num)+" is not a prime number."
    return str(num)+" is a prime number."

num = int(input("Enter number: "))
print(is_prime(num))
```

```python
# Exercise 13: Fibonacci  

def fibonacci(n):
    fib = []
    for i in range(n):
        if i==0 or i==1:
            fib.append(1)
        else:
            fib.append(fib[-2]+fib[-1])
    return fib

n = int(input("Enter number of Fibonacci numbers: "))
print(fibonacci(n))
```

```python
# Exercise 14: List Remove Duplicates 

import random

def duplicates_loop(a):
    new = []
    for x in a:
        if x not in new:
            new.append(x)
    return sorted(new)

def duplicates_set(a):
    return list(set(a))

a = [random.randint(1,10) for i in range(20)]
print(duplicates_loop(a))
print(duplicates_set(a))
```

```python
# Exercise 18: Cows And Bulls

import random

def cows_and_bulls():
    num = str(random.randint(1000,10000))
    guesses = 0
    while True:
        g = input("Guess the 4-digit number: ")
        guesses += 1
        cows = 0
        bulls = 0
        for i in range(len(num)):
            if num[i]==g[i]:
                cows += 1
            elif g[i] in num:
                bulls += 1
        if cows==4:
            return "You guessed right on the {}th try!".format(guesses)
        else:
            print("You found {} cows and {} bulls.".format(cows, bulls))

print(cows_and_bulls())
```

```python
# Exercise 20: Element Search

def search_simple(a, n):
    return n in a

def search_binary(a,n):
    while True:
        if n==a[len(a)//2]:
            return True
        elif len(a)==1:
            return False
        elif n<a[len(a)//2]:
            a = a[:len(a)//2]
        else:
            a = a[len(a)//2:]
            
a = [1, 3, 5, 30, 42, 43, 500]
n = 4
print(search_simple(a,n))
print(search_binary(a,n))
```

```python
# Exercise 27: Tic Tac Toe Draw

def tic_tac_toe_draw():
    board = [[0 for i in range(3)] for j in range(3)]
    moves = 0
    pieces = {1:"X",2:"O"}
    while moves<9:
        if moves%2==0:
            current_player = 1
        else:
            current_player = 2
        row, col = map(lambda x: int(x)-1, input("Player {} picks (row,column): ".format(current_player)).split(","))
        if board[row][col]!=0:
            print("Position already filled.")
        else:
            board[row][col] = pieces[current_player]
            moves += 1
    print("Game finished.")
    return board

print(tic_tac_toe_draw())
```

```python
# Exercise 28: Max Of Three

import random

def max_of_three(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    else:
        return c

a,b,c = random.sample(range(100), 3)
print(max_of_three(a,b,c))
```

```python
# Exercise 29: Tic Tac Toe Game

def draw_board(board):
    for i in range(3):
        print(" ---"*3+" ")
        print("| {} | {} | {} |".format(board[i][0],board[i][1],board[i][2]))
    print(" ---"*3+" ")
    
def check_winner(board):
    winner=0
    if board[0][0]==board[0][1]==board[0][2]!=0 or board[0][0]==board[1][0]==board[2][0]!=0:
        winner = board[0][0]
    elif board[1][0]==board[1][1]==board[1][2]!=0 or board[0][1]==board[1][1]==board[2][1]!=0 or \
    board[0][0]==board[1][1]==board[2][2]!=0 or board[2][0]==board[1][1]==board[2][0]!=0:
        winner = board[1][1]
    elif board[2][0]==board[2][1]==board[2][2]!=0 or board[0][2]==board[1][2]==board[2][2]!=0:
        winner = board[2][2]
    return winner

def tic_tac_toe():
    moves = 0
    pieces = {1:"X",2:"O"}
    board = [[0 for i in range(3)] for j in range(3)]
    draw_board(board)
    while moves<9:
        if moves%2==0:
            current_player = 1
        else:
            current_player = 2
        row, col = map(lambda x: int(x)-1, input("Player {} picks (row,column): ".format(current_player)).split(","))
        while board[row][col]!=0:
            print("Position already filled.")
            row, col = map(lambda x: int(x)-1, input("Player {} picks (row,column): ".format(current_player)).split(","))
        board[row][col] = pieces[current_player]
        moves += 1
        draw_board(board)
        winner = check_winner(board)
        if winner in ["X","O"]:
            print("Player {} wins!".format(current_player))
            return current_player
    print("The game is a tie.")
    return 0

if __name__=='__main__':
    score = [0,0]
    while True:
        winner = tic_tac_toe()
        if winner!=0:
            score[winner-1] += 1
        print("SCORE")
        print("Player 1:",score[0])
        print("Player 2:",score[1])
        if input("Play another game? (yes/no) ")=="no":
            break
```

```python
# Exercise 31: Guess Letters

def guess_letters(word):
    word = list(word)
    guessed = ['_']*len(word)
    while guessed!=word:
        guess = input("Guess a letter: ")
        if guess in word:
            if guess in guessed:
                print("Letter already guessed!")
            else:
                for i in range(len(word)):
                    if word[i]==guess:
                        guessed[i] = guess
                print(*guessed, sep=' ')
        else:
            print("{} not in word".format(guess))
            
guess_letters('EVAPORATE')
```

```python
# Exercise 32: Hangman

import random

def pick_word():
    with open('sowpods.txt') as file:
        words = file.readlines()
    return random.choice(words)

def show_hangman(failed):
    hanged = ['\n|~~~~\n|   |\n|   O\n|\n|\n|', '\n|~~~~\n|   |\n|   O\n|   |\n|\n|', 
              '\n|~~~~\n|   |\n|   O\n|  /|\n|\n|', '\n|~~~~\n|   |\n|   O\n|  /|\\\n|\n|', 
              '\n|~~~~\n|   |\n|   O\n|  /|\\\n|  /\n|', '\n|~~~~\n|   |\n|   O\n|  /|\\\n|  / \\\n|']
    print(*hanged[failed-1])    

def guess_letters(word):
    word = list(word)
    guessed = ['_']*len(word)
    letters_guessed = set()
    failed = 0
    while guessed!=word and failed<6:
        guess = input("Guess a letter: ")
        if guess in letters_guessed:
            print("Letter already guessed!")
        elif guess in word:
            for i in range(len(word)):
                if word[i]==guess:
                    guessed[i] = guess
            print(*guessed, sep=' ')
        else:
            print("{} not in word".format(guess))
            failed += 1
            letters_guessed.add(guess)
            print(*guessed, sep=' ')
            show_hangman(failed)
    if guessed==word:
        print('You Win!')
    else:
        print("You're hanged...")

        
if __name__=='__main__':
    play = "yes"
    while play=="yes":
        word = pick_word()
        guess_letters(word)
        play = input("Play another game? (yes/no) ")
```

```python
# Exercise 36: Birthday Plots

from bokeh.plotting import figure, show, output_file
import json
from collections import Counter
import calendar

with open('scientist_birthdays.json') as file:
    birthdays = json.load(file)
    
months = []
for name, bday in birthdays.items():
    months.append(calendar.month_name[int(bday.split('/')[0])])
    counts = dict(Counter(months))
    
output_file('plot.html')
x_categories = calendar.month_name[1:]
x = list(counts.keys())
y = list(counts.values())
p = figure(x_range=x_categories)
p.vbar(x=x, top=y, width=0.5)
show(p)
```
